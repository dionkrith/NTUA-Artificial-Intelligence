Το σύστημά σας πρέπει να προβλέπει αν μια ταινία θα πρέπει να προταθεί στον χρήστη. Μια ταίνια θα έπρεπε να έχει προταθεί στον  χρήστη, αν έχει βαθμό μεγαλύτερο του 3. Στην συνέχεια, σε συνδιασμό με τις πραγματικές απαντήσεις του χρήστη θα αξιολογήσετε το σύστημά σας χρησιμοποιόντας τις μετρικές: precision, recall, f1 οι οποίες είναι οι πλέον γνωστές μετρικές και ευρέως χρησιμοποιούμενες τεχνικές για την επίβλεψη-μέτρηση απόδοσης ανάλογων συστημάτων.

1. Precision: Δείχνει πόσο ακριβές είναι το σύστημα. Υπολογίζει πόσα από τα στιγμιότυπα τα όποια προβλέψαμε ότι ανήκουν σε μια κλάση όντως ανήκουν σε αυτή. Η μετρική αυτή μας δίνει μια εικόνα σχετικά με τον αριθμό των ταινίων που προβλέψαμε ως προτεινόμενες ενώ δεν θα έπρεπε.

2. Recall: Υπολογίζει πόσα από τα στιγμιότυπα που ανήκουν σε μια κλάση (π.χ. προτεινόμενες ταινίες) προβλέφθηκαν σωστά.

3. F1: Είναι ένας μέσος μεταξύ των παραπάνω δυο μετρικών, έτσι ώστε να διατηρείται μια ισορροπία μεταξύ τους. Υπολογίζεται από την παρακάτω σχέση:

$$F_1=2\frac{Precision\times{Recall}}{Precision+Recall}$$

# **Μέρος 1: Μελέτη των Metadata, Δημιουργία κόσμου και των βασικών queries.**


````python
import pandas as pd
from pyswip import Prolog
````

````python
# Η βιβλιοθήκη pandas είναι χρήσιμη για την εργασία με τέτοια δεδομένα
import pandas as pd 
# Διάβασμα του αρχείου 'movie_metadata.csv' 
data = pd.read_csv("movies_metadata.csv") 
#Στο csv υπαρχούν κελία με nan τιμές
#Στις θέσεις αυτές βάζουμε 'UNK' πράγμα που  κάνουμε με την παρακάτω συνάρτηση
data.fillna("UNK", inplace=True)
# Preview the first 5 lines of the loaded data 
data.head()
````

````python
def clean_text(text):
    text = text.replace(u'\xa0', u'')
    text = text.replace(u"'", u'')
    return text
````

### Ερώτημα 1α - Δημιουργία του κόσμου του προβλήματος

````python
#create World
#Ορίζουμε τον κόσμο μας
prolog = Prolog()

#Για κάθε row του πίνακα φτιάχνουμε τα κατηγορήματα που θέλουμε να αποθηκέυσουμε
#αρχικά σε μια λίστα με το όνομα literals
literals = []
movie_score = {}
for row in data.itertuples(index=True, name='Pandas'):
    
    movie_title = clean_text(getattr(row, 'movie_title'))
    identifier = str(getattr(row, 'id'))
    
    literals.append("identifier('"+ identifier +"','"+  movie_title +"')")
    
    literals.append("director('"+ identifier +"','"+  clean_text(getattr(row, 'director_name')) +"')")
    
    for genre in getattr(row, 'genres').split("|"):
        literals.append("genre('"+ identifier +"','"+ genre +"')")
    
    literals.append("actor('"+ identifier +"','"+  clean_text(getattr(row, 'actor_1_name')) +"')")
    literals.append("actor('"+ identifier +"','"+  clean_text(getattr(row, 'actor_2_name')) +"')")
    literals.append("actor('"+ identifier +"','"+  clean_text(getattr(row, 'actor_3_name')) +"')")
    
    
    flag = 1
    for plot in getattr(row, 'plot_keywords').split("|"):
        if plot == 'black and white':
            literals.append("black_and_white('"+ identifier +"')")
            flag = 0
        else :
            literals.append("plot('"+ identifier +"','"+ clean_text(plot) +"')")
    
    if flag == 1:
        literals.append("colored('"+ identifier +"')")
        
    
    literals.append("language('"+ identifier +"','"+  getattr(row, 'language') +"')")
    
       
    title_year = getattr(row, 'title_year')
    if title_year != 'UNK':
        if int(title_year) < 2000 :
            literals.append("old_movie('"+ identifier +"')")
        elif int(title_year) < 2010 :
            literals.append("semi_modern_movie('"+ identifier +"')")
        else :
            literals.append("modern_movie('"+ identifier +"')")
   
    duration = getattr(row, 'duration')
    if duration != 'UNK':
        if int(duration) < 40 :
            literals.append("sort_movie('"+ identifier +"')")
        elif int(duration) < 220 :
            literals.append("normal_duration_movie('"+ identifier +"')")
        else :
            literals.append("long_movie('"+ identifier +"')")            
    
    
    budget = getattr(row, 'budget')
    if int(budget) > 38000000 :
        literals.append("high_budget('"+ identifier +"')")
    else :
        literals.append("low_budget('"+ identifier +"')") 
        
    production_companies = getattr(row, 'production_companies')
    production_companies = production_companies.split('}')
    for p in production_companies[:len(production_companies) - 1]:
        literals.append("production_company('"+ identifier +"','"+ p[-1] +"')") 
        
    literals.append("country('"+ identifier +"','"+  clean_text(getattr(row, 'country')) +"')")
    

#Η Prolog θέλει τα κατηγορήματά της με την σειρά 
literals.sort()
for literal in literals:
    prolog.assertz(literal)

#Επίσης μπορούμε να κάνουμε consult ένα έτοιμο αρχείο στον κόσμο όπως παρακάτω
#prolog.consult("db.pl")
````

### Ορισμός recommender

Έπειτα, ορίζουμε τη συνάρτηση <i>simple_recommender</i> η οποία δέχεται ως όρισμα μια ταινία και επιστρέφει ένα dictionary με keys τις τιμές από 0 έως 9 και values σύνολα με ταινίες όμοιες με την ταινία (που έχει περαστεί ως όρισμα στη συνάρτηση) σε επίπεδο από 1 έως 10 αντίστοιχα (κάθε ταινία εντάσσεται μια φορά στο dictionary στο μεγαλύτερο επίπεδο που μπορεί).

````python
def recommender(movie):
    
    answers = {i : set() for i in range(0, 10)}
    s = set()
    
    
    q = prolog.query("find_sim_10('" + clean_text(movie) +"', M)")
    for soln in q:
        m = soln['M'] 
        if m not in s:
            s.add(soln['M'])
            answers[9].add(soln['M'])
    q.close()    
    
    
    q = prolog.query("find_sim_9('" + clean_text(movie) +"', M)")
    for soln in q:
        m = soln['M'] 
        if m not in s:
            s.add(soln['M'])
            answers[8].add(soln['M'])
    q.close()
    
    
    q = prolog.query("find_sim_8('" + clean_text(movie) +"', M)")
    for soln in q:
        m = soln['M'] 
        if m not in s:
            s.add(soln['M'])
            answers[7].add(soln['M'])
    q.close()    
    
    
    q = prolog.query("find_sim_7('" + clean_text(movie) +"', M)")
    for soln in q:
        m = soln['M'] 
        if m not in s:
            s.add(soln['M'])
            answers[6].add(soln['M'])
    q.close()
    
    
    q = prolog.query("find_sim_6('" + clean_text(movie) +"', M)")
    for soln in q:
        m = soln['M'] 
        if m not in s:
            s.add(soln['M'])
            answers[5].add(soln['M'])
    q.close()    
    
    
    q = prolog.query("find_sim_5('" + clean_text(movie) +"', M)")
    for soln in q:
        m = soln['M'] 
        if m not in s:
            s.add(soln['M'])
            answers[4].add(soln['M'])
    q.close()

    
    q = prolog.query("find_sim_4('" + clean_text(movie) +"', M)")
    for soln in q:
        m = soln['M'] 
        if m not in s:
            s.add(soln['M'])
            answers[3].add(soln['M'])
    q.close()    
    
    
    q = prolog.query("find_sim_3('" + clean_text(movie) +"', M)")
    for soln in q:
        m = soln['M'] 
        if m not in s:
            s.add(soln['M'])
            answers[2].add(soln['M'])
    q.close()    

    q = prolog.query("find_sim_2('" + clean_text(movie) +"', M)")
    for soln in q:
        m = soln['M'] 
        if m not in s:
            s.add(soln['M'])
            answers[1].add(soln['M'])
    q.close()    
    
    q = prolog.query("find_sim_1('" + clean_text(movie) +"',M)")
    for soln in q:
        m = soln['M'] 
        if m not in s:
            s.add(soln['M'])
            answers[0].add(soln['M'])
    q.close()    


    
    return answers
````

# **3ο Μέρος: Recommendation System Με βάση τις προτιμήσεις-Αξιολογήσεις του χρήστη-Εκπαίδευση και Πρόβλεψη**


Αρχικά μελετάμε τις αξιολογήσεις κάθε χρήστη για να καταλάβουμε την δομή και τις πληροφορίες κάθε αρχείου.




Μέσω της συνάρτησης train_recommender εκπαιδεύουμε τον recommender με βάση το train set, δηλαδή τις αξιολογήσεις ενός συγκεκριμένου χρήστη. Πιο αναλυτικά, για όλες τις ταινίες που αυτος έχει αξιολογήσει (με βαθμολογία από 0 έως 5) αντιστοιχίζουμε στο recommender ένα βάρος σε κάθε όμοια ταινία με αυτή, με βάση τη βαθμολογία που έβαλε ο συγκεκριμένος χρήστης και το επίπεδο ομοιότητας στο οποίο οι δύο ταινίες εντάσσονται. Έτσι, για ταινίες που ο χρήστης έχει βαθμολογήσει με 0 ή 1 επιλέγουμε αρνητικά βάρη, καθώς θεωρούμε ότι ο χρήστης προτιμάει και προτίνουμε σε αυτόν ταινίες όμοιες με ταινία που έχει βαθμολογήσει με 2 και πάνω, ενώ παράλληλα όσο πιο όμοια είναι μια ταινία με αυτή τόσο περισσότερο αυξάνουμε το θετικό (βαθμολογία από 2 έως 5) ή αρνητικό της βάρος (βαθμολογία από 0 έως 1). Με τον τρόπο αυτό μετά την εκπαίδευση του recommender θα προτίνουμε στο χρήστη ταινίες που έχουν συγκεντρώσει θετικό score (βάρος).  
Έπειτα, μέσω της συνάρτησης predict προβλέπουμε, με βάση τα score για κάθε ταινία που υπολογίσαμε παραπάνω, ποιές από τις ταινίες αυτές θα προτιμούσε ο χρήστης, οπότε και τις προτίνουμε και τελικά χρησιμοποιούμε τη συνάρτηση get_metrics για να δούμε με βάση το test set πόσες από τις ταινίες που πρότινε ο recommender συγκεκριμένο χρήστη πράγματι προτιμούνται από αυτόν. Αυτό το επιτυγχάνουμε μέσω των μετρικών precision, recall και F1. 

````python
from tqdm.notebook import tqdm
from sklearn.metrics import precision_score, recall_score, f1_score


rating_weights = {0: -1, 1: -0.5, 2:0, 3:0.3, 4:0.7, 5:1}
score_weights = {i:i + 1 for i in range(0, 8)} # ανάλογα με τα επίπεδα ομοιότητας που έχουν οριστεί στην simple_recommender

def train_recommender(ratings, rating_weights, score_weights, start=0, end = -1, movie_score = {}, simple = 1):
    """
    Στην συνάρτηση αυτή μπορούμε να ορίζουμε ποιο υποσύνολο των αξιολογήσεων θα χρησιμοποιήσουμε για το train,
    ενώ παράλληλα μπορούμε να αρχικοποιήσουμε και το movie_score το οποίο αποθηκεύονται τα score κάθε ταινίας.
    Έτσι αν έχουμε κάνει train τον recommender μας για τις πρώτες 10 ταινίες, μπορούμε να συνεχίσουμε για τις 
    υπόλοιπες χωρίς κάθε φορά να ξανά-διαμορφώνουμε το score για αυτές. Αυτό θα σας βοηθήσει στην εκτέλεση και χρονικά όταν ελέγχετε την 
    απόδοση σε υποσύνολα των ratings π.χ. πρώτες 10 ταινίες έπειτα τις επόμενες 20 κ.ο.κ.
    """
    if end == -1:
        end = len(ratings)

    l = 8
    
    if simple == 0:
        score_weights[8] = 9
        score_weights[9] = 10
        l = 10
        
    ratings = ratings.iloc[range(start, end)]
    for row in tqdm(ratings.itertuples(index=True, name='Pandas')):
        movie = clean_text(getattr(row, 'movie_title'))
        rating = getattr(row, 'rating')

        if simple == 1:
            similar_movies = simple_recommender(movie)
        else :
            similar_movies = recommender(movie)
            
        for i in range(0, l):
            for similar_movie in similar_movies[i]:
                if similar_movie not in movie_score:
                    movie_score[similar_movie] = rating_weights[int(rating)] * score_weights[i]
                else:
                    movie_score[similar_movie] += rating_weights[int(rating)] * score_weights[i] #το weight θα το ορίσετε ανα επίπεδο ομοιότητας οι πολύ όμοιες ταινίες θα έχουν μεγαλύτερο βάρος
    return movie_score


#αυτό είναι ένα παράδειγμα για το πως θα μπορούσε να υλοποιήθει η predict
#έχουμε ορίσει ότι μια ταινία θα έπρεπε να είναι προτεινόμενη αν είχε σκορ > 0
def predict(ratings, movie_score):
    real, pred = [], []
    for i, row in enumerate(ratings.itertuples(index=True, name='Pandas')):
        movie = clean_text(getattr(row, 'movie_title'))
        rating = getattr(row, 'rating')

        if movie in movie_score: #αν έχουμε σχηματίσει βαθμολογία για την ταινία αυτή
            pred.append(int(movie_score[movie] > 0)) #heuristic για το αν μια ταινία είναι προτεινόμενη
            real.append(int(rating > 3))# έτσι ορίζουμε ότι μια ταινία θα έπρεπε να είναι προτεινόμενη
            #η συνθήκη αυτή δεν μπορεί να αλλάξει
        else: #δεν μπορούμε να προτείνουμε κάτι για το όποιο δεν έχουμε σχηματίσει εικόνα
            pred.append(0)
            real.append(int(rating > 3))

    return real, pred


def get_metrics(real, pred):
    metrics = {}
    metrics["precision"] = precision_score(real, pred)
    metrics["recall"] = recall_score(real, pred)
    metrics["f1"] = f1_score(real, pred)
    return metrics
````

````python
def simple_recommender(movie):
    
    answers = {i : set() for i in range(0, 8)}
    s = set()
    
    
    q = prolog.query("simple_find_sim_8('" + clean_text(movie) +"', M)")
    for soln in q:
        m = soln['M'] 
        if m not in s:
            s.add(soln['M'])
            answers[7].add(soln['M'])
    q.close()    
    
    
    q = prolog.query("simple_find_sim_7('" + clean_text(movie) +"', M)")
    for soln in q:
        m = soln['M'] 
        if m not in s:
            s.add(soln['M'])
            answers[6].add(soln['M'])
    q.close()
    
    
    q = prolog.query("simple_find_sim_6('" + clean_text(movie) +"', M)")
    for soln in q:
        m = soln['M'] 
        if m not in s:
            s.add(soln['M'])
            answers[5].add(soln['M'])
    q.close()    
    
    
    q = prolog.query("simple_find_sim_5('" + clean_text(movie) +"', M)")
    for soln in q:
        m = soln['M'] 
        if m not in s:
            s.add(soln['M'])
            answers[4].add(soln['M'])
    q.close()

    
    q = prolog.query("simple_find_sim_4('" + clean_text(movie) +"', M)")
    for soln in q:
        m = soln['M'] 
        if m not in s:
            s.add(soln['M'])
            answers[3].add(soln['M'])
    q.close()    
    
    
    q = prolog.query("simple_find_sim_3('" + clean_text(movie) +"', M)")
    for soln in q:
        m = soln['M'] 
        if m not in s:
            s.add(soln['M'])
            answers[2].add(soln['M'])
    q.close()    

    q = prolog.query("simple_find_sim_2('" + clean_text(movie) +"', M)")
    for soln in q:
        m = soln['M'] 
        if m not in s:
            s.add(soln['M'])
            answers[1].add(soln['M'])
    q.close()    
    
    q = prolog.query("simple_find_sim_1('" + clean_text(movie) +"',M)")
    for soln in q:
        m = soln['M'] 
        if m not in s:
            s.add(soln['M'])
            answers[0].add(soln['M'])
    q.close()    


    
    return answers
````
